```python
import json
from random import randint
```

```python
with open("result.json") as f:
    data = json.load(f)
```

```python
def prev_space(idx, inp_str):
    for i in range(idx, -1, -1):
        if inp_str[i] == ' ':
            return i
    return 0
```

```python
for deal in data:
    if '%' in deal["description"] and 'up to' in deal["description"].lower():
        estimate = {
            "type": 'PO',
        }
        starting_idx = deal["description"].find('%')
        amount = float(deal["description"][prev_space(starting_idx, deal["description"]):starting_idx])
        deal["deal_estimate"] = amount
```
