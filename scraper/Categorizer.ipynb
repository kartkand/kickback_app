```python
import json
```

```python
with open("full_deals.json") as f:
    data = json.load(f)
```

```python
new_data = {}
data
```

```python
uniq_keys = []
for deal in data:
    categories = deal["category"].split(",") if "category" in deal else []
    for c in categories:
        if not c in uniq_keys:
            uniq_keys.append(c)
```

```python
for key in uniq_keys:
    new_data[key] = []
```

```python
for deal in data:
    categories = deal["category"].split(",") if "category" in deal else []
    for c in categories:
        new_data[c].append(deal)
```

```python
with open("full_data_new.json", "w") as fp:
    json.dump(new_data, fp)
```
